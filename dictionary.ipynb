```python

# Multiple users login tracker (simple version)

users = {
    "admin": "1234",
    "user1": "1111",
    "user2": "2222"
}

access = {
    "admin": 0,
    "user1": 0,
    "user2": 0
}

while True:
    username = input("Enter username (or type 'exit' to quit): ")

    if username.lower() == "exit":
        print("Program exited")
        break

    password = input("Enter password: ")

    if username in users and users[username] == password:
        access[username] += 1
        print("Login successful")
        print(username, "accessed system", access[username], "time(s)")
    else:
        print("Invalid login")
```
